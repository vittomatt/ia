```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from fastai.tabular.all import *
```

```python
# Import the CSV and show it
df = pd.read_csv('./titanic.csv');

# Show the columns and types
# print(df.dtypes)

# Show info
df
```

```python
# Do a graph
x_plt = df['Age'].to_numpy()
y_plt = df['Fare'].to_numpy()

plt.plot(x_plt, y_plt, 'r-')
plt.title('X vs Survived')
plt.xlabel('X')
plt.ylabel('Survived')
plt.show()
```

```python
y = ['2urvived']
x = ['Pclass', 'Fare', 'Sex', 'Age', 'sibsp', 'Parch']

scaler = MinMaxScaler()

# Transformations
fare = scaler.fit_transform(df[['Fare']])
age = scaler.fit_transform(df[['Age']])
sibsp = scaler.fit_transform(df[['sibsp']])
Parch = scaler.fit_transform(df[['Parch']])
pclass = pd.get_dummies(df['Pclass'], prefix='pclass')
# ['pclass_1', 'pclass_2']

# Replace the old columns and add the new ones
df['Fare'] = fare
df['Age'] = age
df['sibsp'] = sibsp
df['Parch'] = Parch
df['Pclass_1'] = pclass['pclass_1']
df['Pclass_2'] = pclass['pclass_2']

x = ['Pclass_1', 'Pclass_2', 'Fare', 'Sex', 'Age', 'sibsp', 'Parch']

print(df[x + y])
```

```python
cont_names, cat_names = ['Fare', 'sibsp', 'Parch'], ['Pclass_1', 'Pclass_2', 'Sex']
y_names = y
valid_idx = list(
    range(
        int(len(df)*0.8),
        len(df)
    )
)

dls = TabularDataLoaders.from_df(
    df=df[x + y],
    procs=[Categorify, FillMissing, Normalize],
    cont_names=cont_names,
    cat_names=cat_names, 
    y_names=y_names, 
    valid_idx=valid_idx,
    bs=64
)

learn = tabular_learner(dls, metrics=accuracy)
learn.fit_one_cycle(10)
```
